```python
df['财富集中度等级'].value_counts()
#数据特征构造
import re

def extract_wealth_level_num(level_str):
    # 正则提取文本中所有的数字
    nums = re.findall(r'\d+', str(level_str))
    nums = [float(num) for num in nums]

    if len(nums) == 1:
        # 单边边界（<3%、≥40%），直接返回这个数字
        return nums[0]
    elif len(nums) == 2:
        # 区间（3-10%），返回中值
        return (nums[0] + nums[1]) / 2
    else:

        return pd.NA

# 应用到数据集
df['财富集中度数值'] = df['财富集中度等级'].apply(extract_wealth_level_num)
df['财富集中度'] = df['财富集中度数值'].round(1).astype(str) + "%"

#删除之前列
df.drop(columns='财富集中度等级', inplace=True)

# 验证结果
print("转换结果校验：")
print(df[['财富集中度等级', '财富集中度']].drop_duplicates())
```
